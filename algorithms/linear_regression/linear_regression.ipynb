```python
import pandas #for datafeching
from pandas import DataFrame
import matplotlib.pyplot as plt #for visualrepresentation
from sklearn.linear_model import LinearRegression #for regression
```

```python
data=pandas.read_csv("revenue_data.csv")
```

```python
data.describe()
```

```python
X=DataFrame(data,columns=["production_budget"])
y=DataFrame(data,columns=["revenue_generated"])
regression=LinearRegression()
regression.fit(X,y)
```

```python
plt.figure(figsize=(10,6))
plt.title("Prediction_Model ")
plt.plot(X,regression.predict(X),color="red",linewidth=2)
plt.scatter(X,y,alpha=0.3)
plt.xlabel("Budget")
plt.ylabel("Revenue")
plt.show()
```
